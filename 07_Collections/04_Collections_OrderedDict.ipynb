<p>An <em>OrderedDict</em> is a dictionary that remembers the order of the keys that were inserted first. If a new entry overwrites an existing entry, the original insertion position is left unchanged. </p>

<p><strong>Example</strong></p>

<p><sub><strong>Code</strong></sub></p>

<pre><code>&gt;&gt;&gt; from collections import OrderedDict
&gt;&gt;&gt; 
&gt;&gt;&gt; ordinary_dictionary = {}
&gt;&gt;&gt; ordinary_dictionary['a'] = 1
&gt;&gt;&gt; ordinary_dictionary['b'] = 2
&gt;&gt;&gt; ordinary_dictionary['c'] = 3
&gt;&gt;&gt; ordinary_dictionary['d'] = 4
&gt;&gt;&gt; ordinary_dictionary['e'] = 5
&gt;&gt;&gt; 
&gt;&gt;&gt; print ordinary_dictionary
{'a': 1, 'c': 3, 'b': 2, 'e': 5, 'd': 4}
&gt;&gt;&gt; 
&gt;&gt;&gt; ordered_dictionary = OrderedDict()
&gt;&gt;&gt; ordered_dictionary['a'] = 1
&gt;&gt;&gt; ordered_dictionary['b'] = 2
&gt;&gt;&gt; ordered_dictionary['c'] = 3
&gt;&gt;&gt; ordered_dictionary['d'] = 4
&gt;&gt;&gt; ordered_dictionary['e'] = 5
&gt;&gt;&gt; 
&gt;&gt;&gt; print ordered_dictionary
OrderedDict([('a', 1), ('b', 2), ('c', 3), ('d', 4), ('e', 5)])
</code></pre>
