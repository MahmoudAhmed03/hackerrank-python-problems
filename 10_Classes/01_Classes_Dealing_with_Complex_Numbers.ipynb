<p><strong>Concept</strong></p>

<p>Python is a fully object-oriented language like C++, Java, etc. For reading about classes, refer <a href="http://www.diveintopython3.net/iterators.html#defining-classes">here</a>.
<br><br>
Methods with a double underscore before and after their name are considered as built-in methods. They are used by interpreters and are generally used in the implementation of overloaded operators or other built-in functionality. <br></p>

<pre>__add__-&gt; Can be overloaded for + operation</pre><br>
<pre>__sub__ -&gt; Can be overloaded for - operation</pre><br>
<pre>__mul__ -&gt; Can be overloaded for * operation</pre>

<p><br><br>
For more information on operator overloading in Python, refer <a href="http://docs.python.org/3.2/reference/datamodel.html">here</a>.</p></div></div></div></div>

```python
import math

class Complex(object):
    
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary
        self.number = complex(real, imaginary)
        self.a = -0.0
        
    def __add__(self, no):
        return f"{(self.number + no.number ).real:0.2f}{(self.number + no.number).imag:+0.2f}i"
        
    def __sub__(self, no):
        return f"{(self.number - no.number ).real:0.2f}{(self.number - no.number).imag:+0.2f}i"
        
        
    def __mul__(self, no):
        return f"{(self.number * no.number ).real:0.2f}{(self.number * no.number).imag:+0.2f}i"

    def __truediv__(self, no):
        return f"{(self.number / no.number ).real:0.2f}{(self.number / no.number).imag + 0.0:+0.2f}i"

    def mod(self):
        return f"{abs(self.number):0.2f}+0.00i"

    def __str__(self):
        if self.imaginary == 0:
            result = "%.2f+0.00i" % (self.real)
        elif self.real == 0:
            if self.imaginary >= 0:
                result = "0.00+%.2fi" % (self.imaginary)
            else:
                result = "0.00-%.2fi" % (abs(self.imaginary))
        elif self.imaginary > 0:
            result = "%.2f+%.2fi" % (self.real, self.imaginary)
        else:
            result = "%.2f-%.2fi" % (self.real, abs(self.imaginary))
        return result

if __name__ == '__main__':
    c = map(float, input().split())
    d = map(float, input().split())
    x = Complex(*c)
    y = Complex(*d)
    print(*map(str, [x+y, x-y, x*y, x/y, x.mod(), y.mod()]), sep='\n')
```
