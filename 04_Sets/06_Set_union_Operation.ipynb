<strong>.union()</strong></p>

<p>The <em>.union()</em> operator returns the union of a set and the set of elements in an iterable. <br>
Sometimes, the <em>|</em> operator is used in place of <em>.union()</em> operator, but it operates only on the set of elements in <em>set</em>.<br> 
Set is immutable to the <em>.union()</em> operation (or <em>|</em> operation).</p>

<p><strong>Example</strong></p>

<pre><code>&gt;&gt;&gt; s = set("Hacker")
&gt;&gt;&gt; print s.union("Rank")
set(['a', 'R', 'c', 'r', 'e', 'H', 'k', 'n'])

&gt;&gt;&gt; print s.union(set(['R', 'a', 'n', 'k']))
set(['a', 'R', 'c', 'r', 'e', 'H', 'k', 'n'])

&gt;&gt;&gt; print s.union(['R', 'a', 'n', 'k'])
set(['a', 'R', 'c', 'r', 'e', 'H', 'k', 'n'])

&gt;&gt;&gt; print s.union(enumerate(['R', 'a', 'n', 'k']))
set(['a', 'c', 'r', 'e', (1, 'a'), (2, 'n'), 'H', 'k', (3, 'k'), (0, 'R')])

&gt;&gt;&gt; print s.union({"Rank":1})
set(['a', 'c', 'r', 'e', 'H', 'k', 'Rank'])

&gt;&gt;&gt; s | set("Rank")
set(['a', 'R', 'c', 'r', 'e', 'H', 'k', 'n'])
</code></pre>
