Today, we're learning about a new data type: <em>sets</em>. </p>

<p><strong>Concept</strong></p>

<p>If the inputs are given on one line separated by a character (the delimiter), use <em>split()</em> to get the separate values in the form of a list.  The delimiter is space (ascii 32) by default.  To specify that comma is the delimiter, use <em>string.split(',').</em>  For this challenge, and in general on HackerRank, space will be the delimiter.</p>

<p>Usage:</p>

<pre><code>&gt;&gt; a = raw_input()
5 4 3 2
&gt;&gt; lis = a.split()
&gt;&gt; print (lis)
['5', '4', '3', '2']
</code></pre>

<p>If the list values are all integer types, use the <em>map()</em> method to convert all the strings to integers.</p>

<pre><code>&gt;&gt; newlis = list(map(int, lis))
&gt;&gt; print (newlis)
[5, 4, 3, 2]
</code></pre>

<p>Sets are an unordered collection of unique values. A single set contains values of any immutable data type. 
<br><br>
<strong>CREATING SETS</strong></p>

<pre><code>&gt;&gt; myset = {1, 2} # Directly assigning values to a set
&gt;&gt; myset = set()  # Initializing a set
&gt;&gt; myset = set(['a', 'b']) # Creating a set from a list
&gt;&gt; myset
{'a', 'b'}
</code></pre>

<p><br></p>

<p><strong>MODIFYING SETS</strong> </p>

<p>Using the <em>add()</em> function:</p>

<pre><code>&gt;&gt; myset.add('c')
&gt;&gt; myset
{'a', 'c', 'b'}
&gt;&gt; myset.add('a') # As 'a' already exists in the set, nothing happens
&gt;&gt; myset.add((5, 4))
&gt;&gt; myset
{'a', 'c', 'b', (5, 4)}
</code></pre>

<p><br>
Using the <em>update()</em> function:</p>

<pre><code>&gt;&gt; myset.update([1, 2, 3, 4]) # update() only works for iterable objects
&gt;&gt; myset
{'a', 1, 'c', 'b', 4, 2, (5, 4), 3}
&gt;&gt; myset.update({1, 7, 8})
&gt;&gt; myset
{'a', 1, 'c', 'b', 4, 7, 8, 2, (5, 4), 3}
&gt;&gt; myset.update({1, 6}, [5, 13])
&gt;&gt; myset
{'a', 1, 'c', 'b', 4, 5, 6, 7, 8, 2, (5, 4), 13, 3}
</code></pre>

<p><br>
<strong>REMOVING ITEMS</strong>
<br></p>

<p>Both the <em>discard()</em> and <em>remove()</em> functions take a single value as an argument and removes that value from the set. If that value is not present, <em>discard()</em> does nothing, but <em>remove()</em> will raise a KeyError exception.</p>

<pre><code>&gt;&gt; myset.discard(10)
&gt;&gt; myset
{'a', 1, 'c', 'b', 4, 5, 7, 8, 2, 12, (5, 4), 13, 11, 3}
&gt;&gt; myset.remove(13)
&gt;&gt; myset
{'a', 1, 'c', 'b', 4, 5, 7, 8, 2, 12, (5, 4), 11, 3}
</code></pre>

<p><br></p>

<p><strong>COMMON SET OPERATIONS</strong>
Using <em>union()</em>, <em>intersection()</em> and <em>difference()</em> functions.
<br></p>

<pre><code>&gt;&gt; a = {2, 4, 5, 9}
&gt;&gt; b = {2, 4, 11, 12}
&gt;&gt; a.union(b) # Values which exist in a or b
{2, 4, 5, 9, 11, 12}
&gt;&gt; a.intersection(b) # Values which exist in a and b
{2, 4}
&gt;&gt; a.difference(b) # Values which exist in a but not in b
{9, 5}
</code></pre>

<p><br>
The <em>union()</em> and <em>intersection()</em> functions are symmetric methods: 
<br></p>

<pre><code>&gt;&gt; a.union(b) == b.union(a)
True
&gt;&gt; a.intersection(b) == b.intersection(a)
True
&gt;&gt; a.difference(b) == b.difference(a)
False
</code></pre>
