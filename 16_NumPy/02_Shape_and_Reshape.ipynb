<p>The <em>shape</em> tool gives a tuple of array dimensions and can be used to change the dimensions of an array.</p>

<p><sub><strong>(a). Using <em>shape</em> to get array dimensions</strong></sub></p>

<pre><code>import numpy

my__1D_array = numpy.array([1, 2, 3, 4, 5])
print my_1D_array.shape     #(5,) -&gt; 1 row and 5 columns

my__2D_array = numpy.array([[1, 2],[3, 4],[6,5]])
print my_2D_array.shape     #(3, 2) -&gt; 3 rows and 2 columns 
</code></pre>

<p><sub><strong>(b). Using <em>shape</em> to change array dimensions</strong></sub>    </p>

<pre><code>import numpy

change_array = numpy.array([1,2,3,4,5,6])
change_array.shape = (3, 2)
print change_array      

#Output
[[1 2]
[3 4]
[5 6]]
</code></pre>

<p><a href="http://docs.scipy.org/doc/numpy/reference/generated/numpy.reshape.html#numpy.reshape"><strong>reshape</strong></a></p>

<p>The <em>reshape</em> tool gives a new shape to an array without changing its data. It creates a new array and does not modify the original array itself. </p>

<pre><code>import numpy

my_array = numpy.array([1,2,3,4,5,6])
print numpy.reshape(my_array,(3,2))

#Output
[[1 2]
[3 4]
[5 6]]
</code></pre>
